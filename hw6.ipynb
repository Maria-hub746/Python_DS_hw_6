```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
dataset = pd.read_csv('data/data_2d.csv')

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init="auto")
    kmeans.fit(dataset)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Кількість кластерів')
plt.ylabel('Сума відстаней до центроїдів')
plt.title('Ліктьова крива')
plt.show()
```

```python
mnist = pd.read_csv('data/mnist.csv')

pca = PCA(n_components=2)
mnist_2d = pca.fit_transform(mnist)

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init="auto")
    kmeans.fit(mnist_2d)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Кількість кластерів')
plt.ylabel('Сума відстаней до центроїдів')
plt.title('Ліктьова крива')
plt.show()

k_optimal = 5

kmeans = KMeans(n_clusters=k_optimal, n_init="auto")
kmeans.fit(mnist_2d)

labels = kmeans.labels_

plt.scatter(mnist_2d[:, 0], mnist_2d[:, 1], c=labels, cmap='viridis')
plt.xlabel('Перша головна компонента')
plt.ylabel('Друга головна компонента')
plt.title('K-means кластеризація для MNIST датасету')
plt.colorbar()
plt.show()
```
